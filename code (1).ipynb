```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Display plots inside Jupyter Notebook
%matplotlib inline
```

```python
df = pd.read_csv("Sample - Superstore.csv", encoding="latin1")

df.head()
```

```python
print("Dataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
```

```python
df.info()
```

```python
df.describe()
```

```python
print("Missing Values")
print(df.isnull().sum())

print("\nDuplicate Rows")
print(df.duplicated().sum())
```

```python
sales = df["Sales"].to_numpy()

profit = df["Profit"].to_numpy()

quantity = df["Quantity"].to_numpy()

print("Sales Array")
print(sales[:10])

print("\nProfit Array")
print(profit[:10])

print("\nQuantity Array")
print(quantity[:10])
```

```python
print("=" * 50)
print("NUMPY ANALYSIS")
print("=" * 50)

print("Total Sales :", np.sum(sales))

print("Total Profit :", np.sum(profit))

print("Total Quantity :", np.sum(quantity))

print("Average Sales :", np.mean(sales))

print("Median Sales :", np.median(sales))

print("Maximum Sale :", np.max(sales))

print("Minimum Sale :", np.min(sales))

print("Standard Deviation :", np.std(sales))
```

```python
print("First 10 Sales")
print(sales[:10])

print("\nLast 10 Sales")
print(sales[-10:])

print("\nSales Greater Than 500")
print(sales[sales > 500])

print("\nSorted Sales")
print(np.sort(sales))

print("\nTop 20 Highest Sales")
print(np.sort(sales)[-20:])
```

```python
print("=" * 50)
print("BUSINESS INSIGHTS")
print("=" * 50)

print("Number of Orders :", len(df))

print("Number of Categories :", df["Category"].nunique())

print("Number of Sub-Categories :", df["Sub-Category"].nunique())

print("Number of Regions :", df["Region"].nunique())

print("Number of Customers :", df["Customer Name"].nunique())

print("Number of States :", df["State"].nunique())

print("\nUnique Categories")
print(df["Category"].unique())

print("\nUnique Regions")
print(df["Region"].unique())
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Month"] = df["Order Date"].dt.strftime("%B")

df["Month_Number"] = df["Order Date"].dt.month

df.head()
```

```python
monthly_sales = (
    df.groupby(["Month_Number", "Month"])["Sales"]
      .sum()
      .reset_index()
      .sort_values("Month_Number")
)

monthly_sales
```

```python
region_sales = (
    df.groupby("Region")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

region_sales
```

```python
category_sales = (
    df.groupby("Category")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

category_sales
```

```python
highest_profit = df.loc[df["Profit"].idxmax()]

lowest_profit = df.loc[df["Profit"].idxmin()]

print("Highest Profit Order")
display(highest_profit)

print("\nLowest Profit Order")
display(lowest_profit)
```

```python
top_products = (
    df.groupby("Sub-Category")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_products
```

```python
plt.figure(figsize=(8,5))

plt.bar(category_sales.index, category_sales.values)

plt.title("Sales by Category")

plt.xlabel("Category")

plt.ylabel("Sales")

plt.grid(axis="y")

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.plot(
    monthly_sales["Month"],
    monthly_sales["Sales"],
    marker="o"
)

plt.title("Monthly Sales Trend")

plt.xlabel("Month")

plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(7,7))

plt.pie(
    region_sales.values,
    labels=region_sales.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Region-wise Sales Distribution")

plt.show()
```

```python
print("=" * 60)
print("SALES DASHBOARD SUMMARY")
print("=" * 60)

print("Total Sales        :", round(df["Sales"].sum(), 2))
print("Total Profit       :", round(df["Profit"].sum(), 2))
print("Average Sales      :", round(df["Sales"].mean(), 2))
print("Average Profit     :", round(df["Profit"].mean(), 2))

print("\nBest Category")
print(category_sales.idxmax())

print("\nBest Region")
print(region_sales.idxmax())

print("\nTop Selling Sub-Category")
print(top_products.idxmax())

print("=" * 60)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(df["Sales"], df["Profit"])

plt.title("Sales vs Profit")

plt.xlabel("Sales")

plt.ylabel("Profit")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=20)

plt.title("Sales Distribution")

plt.xlabel("Sales")

plt.ylabel("Frequency")

plt.grid(True)

plt.show()
```

```python
quantity_analysis = (
    df.groupby("Category")["Quantity"]
      .sum()
      .sort_values(ascending=False)
)

quantity_analysis
```

```python
profit_category = (
    df.groupby("Category")["Profit"]
      .sum()
      .sort_values(ascending=False)
)

profit_category
```

```python
top_customers = (
    df.groupby("Customer Name")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_customers
```

```python
top_states = (
    df.groupby("State")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_states
```

```python
plt.figure(figsize=(8,5))

plt.bar(category_sales.index, category_sales.values)

plt.title("Sales by Category")

plt.xlabel("Category")

plt.ylabel("Sales")

plt.tight_layout()

plt.savefig("dashboard.png")

plt.show()

print("Dashboard saved successfully!")
```

```python
summary = pd.DataFrame({

    "Metric":[
        "Total Sales",
        "Total Profit",
        "Average Sales",
        "Average Profit",
        "Maximum Sale",
        "Minimum Sale"
    ],

    "Value":[
        df["Sales"].sum(),
        df["Profit"].sum(),
        df["Sales"].mean(),
        df["Profit"].mean(),
        df["Sales"].max(),
        df["Sales"].min()
    ]

})

summary
```

```python
summary.to_csv("sales_summary.csv", index=False)

print("Summary CSV Saved Successfully!")
```
